## 사용한 Data는 HAM10000
출처: https://dataverse.harvard.edu/dataset.xhtml?persistentId=doi:10.7910/DVN/DBW86T  
   
아래의 2개의 zip 파일의 이미지를 사용함  

- HAM10000_images_part1.zip (5000 JPEG files)  
- HAM10000_images_part2.zip (5015 JPEG files)  
(용량의 제약으로 사용한 이미지들을 git에 업로드하지 못함)

```python
import os
file_list = os.listdir('HAM10000_images')   # github에는 없는 폴더
file_list
```

```python
# 멜라닌세포 모반 (nv)
# 흑색종 (mel)

nv_list = metaData[metaData['dx'] == 'nv']['image_id'].values.tolist()
mel_list = metaData[metaData['dx'] == 'mel']['image_id'].values.tolist()
```

```python
# nv_image에는 멜라닌세포 모반 이미지만
# mel_image에는 흑색종 이미지만

file_source = 'HAM10000_images'
file_mel = 'mel_image'
file_nv = 'nv_image'
 
get_files = os.listdir(file_source)
 
for image in mel_list:
    os.replace(file_source + "\\" + image + ".jpg", file_mel + "\\" + image + ".jpg")
    
for image in nv_list:
    os.replace(file_source + "\\" + image + ".jpg", file_nv + "\\" + image + ".jpg")
```
